```python
import pandas as pd
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
ticker = 'TM'
name = 'toyota'
```

```python
rev = pd.read_html('https://www.macrotrends.net/stocks/charts/'+ticker+'/'+name+'/revenue' ,index_col=0)[1].T
production_costs = pd.read_html('https://www.macrotrends.net/stocks/charts/'+ticker+'/'+name+'/cost-goods-sold' ,index_col=0)[1].T
cash_on_hand = pd.read_html('https://www.macrotrends.net/stocks/charts/'+ticker+'/'+name+'/cash-on-hand' ,index_col=0)[1].T
equity = pd.read_html('https://www.macrotrends.net/stocks/charts/'+ticker+'/'+name+'/total-share-holder-equity' ,index_col=0)[1].T
net_profit = pd.read_html('https://www.macrotrends.net/stocks/charts/'+ticker+'/'+name+'/net-income' ,index_col=0)[1].T
shares = pd.read_html('https://www.macrotrends.net/stocks/charts/'+ticker+'/'+name+'/shares-outstanding' ,index_col=0)[1].T
assets_total = pd.read_html('https://www.macrotrends.net/stocks/charts/'+ticker+'/'+name+'/total-assets' ,index_col=0)[1].T
current_assets = pd.read_html('https://www.macrotrends.net/stocks/charts/'+ticker+'/'+name+'/total-current-assets' ,index_col=0)[1].T
current_debt = pd.read_html('https://www.macrotrends.net/stocks/charts/'+ticker+'/'+name+'/total-current-liabilities' ,index_col=0)[1].T
supplies = pd.read_html('https://www.macrotrends.net/stocks/charts/'+ticker+'/'+name+'/inventory' ,index_col=0)[1].T
long_loans = pd.read_html('https://www.macrotrends.net/stocks/charts/'+ticker+'/'+name+'/long-term-debt' ,index_col=0)[1].T

for i in range(len(rev.columns)):
    rev[str(rev.columns[i])] = rev[str(rev.columns[i])].str.replace('$','').str.replace(',','')
    production_costs[str(production_costs.columns[i])] = production_costs[str(production_costs.columns[i])].str.replace('$','').str.replace(',','')
    net_profit[str(net_profit.columns[i])] = net_profit[str(net_profit.columns[i])].str.replace('$','').str.replace(',','')
    cash_on_hand[str(cash_on_hand.columns[i])] = cash_on_hand[str(cash_on_hand.columns[i])].str.replace('$','').str.replace(',','')
    equity[str(equity.columns[i])] = equity[str(equity.columns[i])].str.replace('$','').str.replace(',','')
    assets_total[str(assets_total.columns[i])] = assets_total[str(assets_total.columns[i])].str.replace('$','').str.replace(',','')
    current_assets[str(current_assets.columns[i])] = current_assets[str(current_assets.columns[i])].str.replace('$','').str.replace(',','')
    current_debt[str(current_debt.columns[i])] = current_debt[str(current_debt.columns[i])].str.replace('$','').str.replace(',','')
    supplies[str(supplies.columns[i])] = supplies[str(supplies.columns[i])].str.replace('$','').str.replace(',','')
    long_loans[str(long_loans.columns[i])] = long_loans[str(long_loans.columns[i])].str.replace('$','').str.replace(',','')
    
rev = rev.apply(pd.to_numeric, errors='ignore')
production_costs = production_costs.apply(pd.to_numeric, errors='ignore')
net_profit = net_profit.apply(pd.to_numeric, errors='ignore')
shares = shares.apply(pd.to_numeric, errors='ignore') 
cash_on_hand = cash_on_hand.apply(pd.to_numeric, errors='ignore') 
equity = equity.apply(pd.to_numeric, errors='ignore') 
assets_total = assets_total.apply(pd.to_numeric, errors='ignore') 
current_assets = current_assets.apply(pd.to_numeric, errors='ignore') 
current_debt = current_debt.apply(pd.to_numeric, errors='ignore') 
supplies = supplies.apply(pd.to_numeric, errors='ignore')
long_loans = long_loans.apply(pd.to_numeric, errors='ignore')

df = pd.concat([rev, production_costs, net_profit, shares, cash_on_hand,
                equity, assets_total, current_assets, current_debt, supplies, long_loans], ignore_index=False) 
df.columns = df.columns.str.slice(start=2 , stop=7)
columns = df.columns.tolist()
columns = columns[::-1]
df = df[columns]
df = df.drop(df.columns[0:-28],axis=1)
df
```

```python
class ZIS:
    def __init__(self , df):
        self.time = df.columns
        self.shares = df.iloc[2]
        self.year = [ self.time[i] for i in range(3,len(self.time),4)]
        #self.year = np.append(self.year , '20/Q4')
        self.revenue = df.iloc[0]
        self.Production_cost = df.iloc[1]
        self.net_profit = df.iloc[2]
        self.Last4_profit = []
        self.Last4_revenue = []
        self.Revenue_year = [ np.sum(self.revenue[i:i+4]) for i in range(0,len(self.shares),4) ]
        self.Profit_year = [np.sum(self.net_profit[i:i+4]) for i in range(0,len(self.shares),4) ]
        self.Last4_profit = [ np.sum(self.net_profit[i-4:i]) for i in range(4,len(self.shares)+1) ]
        self.Last4_revenue = [ np.sum(self.revenue[i-4:i]) for i in range(4,len(self.shares)+1) ]
            
        self.Current_assets = df.iloc[7]
        self.cash = df.iloc[4]
        self.Assets_total = df.iloc[6]
        self.Short_debt = df.iloc[8]
        self.Supplies = df.iloc[9]
        self.Loans = df.iloc[10]
        
        self.Assets_year_mean = [ np.mean(self.Assets_total[i:i+4]) for i in range(0,len(self.shares),4) ]
        self.Supplies_year_mean = [ np.mean(self.Supplies[i:i+4]) for i in range(0,len(self.shares),4) ]
        self.Equity = df.iloc[5]
        self.beta = self.Equity[3:len(self.Equity)]/self.Last4_revenue
        self.ROE = self.Last4_profit/self.Equity[3:len(self.Equity)]
        self.Assets_productivity = [ self.Revenue_year[i]/self.Assets_year_mean[i]*100 for i in range(0,len(self.Revenue_year))]
        self.Supply_productivity = [ self.Profit_year[i]/self.Supplies_year_mean[i]*100 for i in range(0,len(self.Revenue_year))]
        self.Current_ratio = self.Current_assets/self.Short_debt
        self.Cash_cover = self.cash/self.Short_debt
        self.SupplyCover = self.Supplies/self.Short_debt
        self.ROI = (self.cash+self.Supplies)/self.Loans
        
    def equity(self):
        self.fig = plt.figure(figsize = (20,15) , dpi=80)
        self.ax = plt.subplot(311)
        self.ax.spines["top"].set_visible(False)
        self.ax.spines["right"].set_visible(False)
        self.ax.yaxis.grid(True)
        self.ax.set_axisbelow(True)
        
        plt.bar(self.time, self.Equity/self.shares*1000, 
                label='equity/share ='"$"+str(round(self.Equity[len(self.shares)-1]/self.shares[len(self.shares)-1],2))+"$")
        plt.plot(self.time, self.cash/self.shares*1000 , color='green', linewidth = 3, 
                 label='cash/share ='"$"+str(round(self.cash[len(self.shares)-1]/self.shares[len(self.shares)-1],2))+"$")
        
        plt.xticks(rotation = 50 , fontsize = 12)
        plt.legend(loc='best' , fontsize = 12 )
        self.ax = plt.subplot(312)
        self.ax.spines["top"].set_visible(False)
        self.ax.spines["right"].set_visible(False)
        self.ax.yaxis.grid(True)
        self.ax.set_axisbelow(True)
        
        plt.bar(self.time[3:len(self.shares)], self.ROE*100, color='darkgreen')
        self.ax.set_title('ROE')
        self.ax.set_ylabel('[%]')
        plt.xticks(rotation = 50 , fontsize = 12)
        plt.tight_layout(pad=3)
        self.ax = plt.subplot(313)
        self.ax.spines["top"].set_visible(False)
        self.ax.spines["right"].set_visible(False)
        self.ax.yaxis.grid(True)
        self.ax.set_axisbelow(True)
        
        plt.plot(self.time[3:len(self.shares)], self.beta[3:len(self.shares)]*self.ROE*100 )
        self.ax.set_title('ROE*BETA')
        self.ax.set_ylabel('[%]')
        plt.xticks(rotation = 50 , fontsize = 12)
        plt.tight_layout(pad=3)
        
    def Assets(self):
        self.fig = plt.figure(figsize = (20,15) , dpi=80)
        self.ax = plt.subplot(211)
        self.ax.spines["top"].set_visible(False)
        self.ax.spines["right"].set_visible(False)
        self.ax.yaxis.grid(True)
        self.ax.set_axisbelow(True)
        self.ax.set_title('Assets' , fontsize = 20)
        plt.bar(self.time, self.Assets_total/1000, label='Total Assets')
        plt.plot(self.time, self.Current_assets/1000, color='firebrick',label='Current Assets')
        self.ax.set_ylabel('w mln')
        plt.xticks(rotation=50)
        self.ax = plt.subplot(212)
        self.ax.spines["top"].set_visible(False)
        self.ax.spines["right"].set_visible(False)
        self.ax.yaxis.grid(True)
        self.ax.set_axisbelow(True)
        plt.plot(self.year , self.Assets_productivity, label='Assets productivity = revenue/assets')
        plt.plot(self.year , self.Supply_productivity, label = 'Supplies productivity = profit/supplies')
        plt.legend(loc='best')
        self.fig = plt.figure(figsize = (20,15) , dpi=80)
        self.ax = plt.subplot(211)
        self.ax.spines["top"].set_visible(False)
        self.ax.spines["right"].set_visible(False)
        self.ax.yaxis.grid(True)
        self.ax.set_axisbelow(True)
        plt.plot(self.time , (self.Current_assets-self.Short_debt)/self.Assets_total*100, label='(Current assets-Short debt)/Total Assets')
        plt.legend(loc='best' , fontsize = 12 )
        plt.xticks(rotation=50)
        
    def Cover(self):
        self.fig = plt.figure(figsize = (20,15) , dpi=80)
        self.ax = plt.subplot(211)
        self.ax.spines["top"].set_visible(False)
        self.ax.spines["right"].set_visible(False)
        self.ax.yaxis.grid(True)
        self.ax.set_axisbelow(True)
        self.ax.set_title('Liquidity' , fontsize=20)
        plt.plot(self.time, self.Current_ratio, label='Current ratio')
        plt.plot(self.time, self.SupplyCover, label='Supply cover')
        plt.plot(self.time, self.Cash_cover, label='Cash cover')
        plt.legend(loc='best')
        plt.xticks(rotation = 50)
        self.ax = plt.subplot(212)
        self.ax.spines["top"].set_visible(False)
        self.ax.spines["right"].set_visible(False)
        self.ax.yaxis.grid(True)
        self.ax.set_axisbelow(True)
        plt.bar(self.time, self.Supplies/self.Production_cost, label= 'Supply/Production Costs')
        plt.plot(self.time, self.ROI, label='ROI= (Cash+supplies)/Loans')
        # plt.plot(self.time, self.Loans)
        plt.xticks(rotation = 50)
        plt.legend(loc='best')
        
    def Flow(self):
        self.fig = plt.figure(figsize = (20,15) , dpi=80)
        self.ax = plt.subplot(211)
        self.ax.spines["top"].set_visible(False)
        self.ax.spines["right"].set_visible(False)
        self.ax.yaxis.grid(True)
        self.ax.set_axisbelow(True)
        self.ax.set_title('Flow' , fontsize=20)
        # plt.plot(self.time , self.FreeCashFlow/self.shares*1000, marker='o', label='Free Cash Flow')  
        plt.legend(loc='best' , fontsize = 15)
        plt.xticks(rotation = 50)
        self.ax = plt.subplot(212)
        self.ax.spines["top"].set_visible(False)
        self.ax.spines["right"].set_visible(False)
        self.ax.yaxis.grid(True)
        self.ax.set_axisbelow(True)
        # plt.plot(self.time , self.operating_financial_score/self.shares*100, marker='o', label='Operating and financial score')
        plt.legend(loc='best' , fontsize = 15)
        plt.xticks(rotation = 50)
        
        self.fig = plt.figure(figsize = (20,15) , dpi=80)
        self.ax = plt.subplot(311)
        self.ax.spines["top"].set_visible(False)
        self.ax.spines["right"].set_visible(False)
        self.ax.yaxis.grid(True)
        self.ax.set_axisbelow(True)
        plt.plot(self.time, self.Supplies/self.Production_cost*100, label='Supplies/Production cost [%]')
        plt.plot(self.time , self.Short_debt/self.Production_cost*100, label='Short debt/Production cost [%]')
        plt.xticks(rotation=50)
        plt.legend(loc='best' , fontsize=15)
        self.ax = plt.subplot(312)
        self.ax.spines["top"].set_visible(False)
        self.ax.spines["right"].set_visible(False)
        self.ax.yaxis.grid(True)
        self.ax.set_axisbelow(True)
        plt.plot(self.time, self.Supplies/self.Sell_profit*100, label='Supplies/Sell profit %')
        plt.xticks(rotation=50)
        plt.legend(loc='best' , fontsize=15)
        self.ax = plt.subplot(313)
        self.ax.spines["top"].set_visible(False)
        self.ax.spines["right"].set_visible(False)
        self.ax.yaxis.grid(True)
        self.ax.set_axisbelow(True)
        plt.plot(self.time, self.Amortization/self.Supplies*100, label='Amortization/Supplies')
        plt.legend(loc='best' , fontsize=15)
        plt.xticks(rotation = 50)  
```
